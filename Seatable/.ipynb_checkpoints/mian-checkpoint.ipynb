```python
from seatable_api import Account, Base, context
from datetime import datetime

email = '18909175158@163.com'
password = 'sincere1027'
server_url = 'https://cloud.seatable.cn/'
account = Account(email, password, server_url)
account.auth()

# 获取量张表的base
model_base = account.get_base(189828, '配件管理总表')
record_base = account.get_base(189828, '维修记录表')

existing_record_num = record_base.list_rows('后台数据')[-1]['总记录量']  # 获取新增维修记录的数量

# 获取配件管理表中现有的数据，和维修记录表中新增的数据
existing_model_list = record_base.list_rows('锻造车间')
increased_record_list = model_base.list_rows('维修记录')[existing_record_num:]

# 分别创建配件更新、追加列表容器
update_date = []
append_date = []

def averageDays(old_list,new_str):
    '''计算平均故障间隔时间（月）
    以配件管理表中过去故障时间和维修记录中新增故障时间（以两者为参数）
    计算平均值 return int
    '''
    # 限制记录的故障日期的次数
    if len(old_list) > 9:
        old_list = old_list[-9:]
        
    # 计算平均故障间隔
    if len(old_list) > 0:
        last_date = datetime.strptime(new_str,'%Y-%m-%d')
        start_date = datetime.strptime(old_list[0],'%Y-%m-%d')
        average_days = (last_date - start_date).days / len(old_list)
    
    return int(average_days)
    
'''
将新旧记录逐行进行单循环比较
先处理核心配件，原来有则迭代，原来没有则新增行
对于普通配件
若，经清洗后疑似型号一致，并属于同一设备
则认定为同一型号
'''
for increased_record in increased_record_list:
    for existing_model in existing_model_list:  # 单循环取出每一行的dict
        if increased_record['故障(核心)配件型号'] != '' and increased_record[
            '故障(核心)配件型号'] == existing_model['配件型号']:  # 如果有核心配件，且旧表中有同型号则处理
            increased_record['故障(核心)配件型号'] = ''  # 使之为空，防止反复处理
            failure_rate = averageDays(
                existing_model['前10次故障日期'],increased_record['故障日期']) # 计算平均故障间隔
            # 添加入批量更新的数据列表
            update_date.append({
                'row_id': existing_model['_id'],
                'row': {
                    '配件等级': '核心',
                    '最近维修/更换时间': increased_record['故障日期'],
                    '平均故障间隔(天)':str(failure_rate),
                    '前10次故障日期':existing_model['前10次故障日期'][-9:].append(increased_record['故障日期'])
                }
            })
```
